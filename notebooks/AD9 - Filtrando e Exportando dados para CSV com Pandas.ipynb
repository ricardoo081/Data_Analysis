# Microdados do ENEM

Importando o pacote <strong>Pandas</strong> único que será utilizado nessa aula:

<em>Documentação do Pandas</em>: https://pandas.pydata.org/pandas-docs/stable/index.html

```python
import pandas as pd
```

Ler o o conjunto de dados

```python
microdadosEnem = pd.read_csv("brutos/microdados_enem2018/DADOS/MICRODADOS_ENEM_2018.csv", sep=";", encoding='ISO-8859-1')
```

Visualizar a cabeça (head) do DataFrame:

```python
microdadosEnem
```

```python
microdadosEnem.columns.values
```

```python
microdadosEnemSelecionado = microdadosEnem.filter(items=['CO_MUNICIPIO_RESIDENCIA','NO_MUNICIPIO_RESIDENCIA', 
                       'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA', 
                       'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 
                       'NU_NOTA_MT','NU_NOTA_REDACAO'])
```

```python
microdadosEnemSelecionado
```

```python
mediaEnemPorMunicipio = microdadosEnemSelecionado.groupby(['CO_MUNICIPIO_RESIDENCIA','NO_MUNICIPIO_RESIDENCIA', 
                       'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA'], as_index=False).mean()
```

```python
mediaEnemPorMunicipio
```

```python
mediaEnemPorMunicipio.to_csv("mediaEnemPorMunicipio.csv")
```
